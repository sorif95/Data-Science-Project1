```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset
url = "https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv"
df = pd.read_csv(url)

# 2. Create 'Number of Relatives' column
df['Number of Relatives'] = df['SibSp'] + df['Parch']

# 3. Filter data for surviving passengers
survivors = df[df['Survived'] == 1]

# 4. Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(survivors['Number of Relatives'], survivors['Age'], alpha=0.5, c='blue')
plt.title('Surviving Passengers: Age vs Number of Relatives')
plt.xlabel('Number of Relatives (SibSp + Parch)')
plt.ylabel('Age')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
